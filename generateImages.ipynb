```python
def rollH(image, delta):
    "Roll an image sideways"
    image = image.copy()
    xsize, ysize = image.size

    if delta == 0: return image

    part1 = image.crop((0, 0, xsize-delta, ysize))
    part2 = image.crop((xsize-delta, 0, xsize, ysize))
    image.paste(part2, (0, 0, delta, ysize))
    image.paste(part1, (delta, 0, xsize, ysize))

    return image


def rollV(image, delta):
    "Roll an image sideways"
    image = image.copy()
    xsize, ysize = image.size

    if delta == 0: return image

    part1 = image.crop((0, 0, xsize, ysize-delta))
    part2 = image.crop((0, ysize-delta, xsize, ysize))
    image.paste(part2, (0, 0, xsize, delta))
    image.paste(part1, (0, delta, xsize, ysize))

    return image
```

```python
# Создает 136 изображений единицы длиной 8 всех возможных положений
def roll_1(baseImage):
    ID = 0

    for i in range(1,9):
        fileName = './new_images/1/new_'+str(ID)+'.png'
        im2 = rollV(baseImage,i)
        im2.save(fileName)
        ID += 1 
        
        for i in range(1,17):
            fileName = './new_images/1/new_'+str(ID)+'.png'
            im3 = rollH(im2,i)
            im3.save(fileName)
            ID += 1 
```

```python
import os
from random import shuffle
import numpy as np
import shutil
```

```python
classes = os.listdir('./generated/')

# Каталог с набором данных
data_dir = '/home/vlad/DataSience/Numbers/generated/'
# Каталог с данными для обучения
train_dir = './data2.0/train'
# Каталог с данными для проверки
val_dir = './data2.0/val'
# Каталог с данными для тестирования
test_dir = './data2.0/test'
# Часть набора данных для тестирования
test_data_portion = 0.15
# Часть набора данных для проверки
val_data_portion = 0.15
```

```python
def create_directory(dir_name):
    if os.path.exists(dir_name):
        shutil.rmtree(dir_name)
    os.makedirs(dir_name)
    for i in classes:
        os.makedirs(os.path.join(dir_name, i))
        
```

```python
create_directory(train_dir)
create_directory(val_dir)
create_directory(test_dir)
```

```python
def shuffle_images(source_dir):
    
    shuffle_list = []
    classes_length = []
    for c in classes:
        images_list = os.listdir(os.path.join(source_dir, c))
        
        classes_length.append(len(images_list))
        shuffle(images_list)
        shuffle_list.append(images_list)
        print(c,len(images_list))
        
    return shuffle_list, classes_length
#     print(shuffle_list)
```

```python
def copy_images(source_dir, dest_dir, start_index, end_index,shuf_list):
    
     
    
    for i in range(len(classes)):
        for name in shuf_list[i][start_index[i]:end_index[i]]:
            print(os.path.join(source_dir + classes[i], name))
            shutil.copy2(os.path.join(source_dir +'/'+ classes[i], name), 
                        os.path.join(dest_dir, classes[i]))
    
        
```

```python
shuf_list, classes_length = shuffle_images(data_dir)

classes_length = np.array(classes_length)
amounts_train = np.array([int(i * 0.7) for i in classes_length])
amounts_val = np.array([int(i * 0.15) for i in classes_length])
amounts_test = classes_length - amounts_train - amounts_val
print(amounts_train)
print(amounts_val)
print(amounts_test)
```
